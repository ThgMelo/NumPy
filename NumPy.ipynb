## Carregando os dados

```python
import numpy as np
```

```python
np.loadtxt('apples_ts.csv', delimiter=',')
```

Tivemos um erro ao importar os dados, o erro ocorreu porque a primeira coluna contém os nomes (string) de cada cidade.     
Precisamos importar e ignorar a primeira coluna.    
Para isso, iremos passar como um parâmetro os nomes das colunas que queremos importar:

```python
# Queremos a primeira coluna e a ultima (7*12+3)
np.arange(1, (7*12+3), 1)
```

```python
colunas = np.arange(1, (7*12+3), 1)
colunas
```

```python
# passando as colunas como parâmetro
np.loadtxt('apples_ts.csv', delimiter=',', usecols=colunas)
```

```python
dado = np.loadtxt('apples_ts.csv', delimiter=',', usecols=colunas)
```

```python
dado
```
